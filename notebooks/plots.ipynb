```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/data_final.csv")
df_scored = pd.read_csv("../data/modified_data.csv")
```

```python
df_res = pd.merge(df, df_scored, on="cv", how="inner")
```

```python
df_plt = df_res[
    ~df_res["result"].isna() & ~df_res["grade"].isna() & df_res["result"] > 0
][["name", "result", "grade"]].reset_index(drop=True)
```

```python
# plt.bar(df_plt.index, df_plt.result / 20, color='blue', alpha=0.7, label='Prediction')
# plt.bar(df_plt.index, df_plt.grade, color='orange', alpha=0.7, label='Grade')

plt.hist(df_plt["result"] / 20, bins=10, alpha=0.7, color="blue", label="Result")
plt.hist(df_plt["grade"], bins=10, alpha=0.7, color="orange", label="Grade")

plt.xlabel("Grade", fontsize=12)
plt.ylabel("Counts", fontsize=12)
plt.title("Comparison of Prediction and Grade", fontsize=14)

plt.legend()

plt.show()
```

```python
plt.bar(df_plt.index, df_plt.result / 20, color="blue", alpha=0.7, label="Prediction")
plt.bar(df_plt.index, df_plt.grade, color="orange", alpha=0.7, label="Grade")

plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.title("Comparison of Prediction and Grade", fontsize=14)

plt.legend()

plt.show()
```
